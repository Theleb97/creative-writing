# Plotting the novel on the plane

```python
import numpy as np
import matplotlib.pyplot as plt

N = 700

t = np.linspace(-0.5, 0.5, N)
turns = 10
speed = 9

#The first part of the text : spiraling inward, going back home
#taking the time to introspect, a dectructive introspection
#as we can see : the horizon of the black hole has been attained.
X1 = np.cos(2 * np.pi * turns * t) * (0.05 + np.exp(-speed * t**2))
Y1 = np.sin(2 * np.pi * turns * t) * (0.05 + np.exp(-speed * t**2))

#The second part of the text : spiraling outward, going out to the world
#and seeking justice, discovering, fill oneself with energy and strength
#a ragefull energy that destroys without care.
X2 = np.cos(2 * np.pi * turns * t + np.pi) * (0.05 + np.exp(-speed * t**2))
Y2 = np.sin(2 * np.pi * turns * t + np.pi) * (0.05 + np.exp(-speed * t**2))


# Both parts are linked by their extremities. 
#The end of the first leads to the begining of the second
#and vice-versa. 
plt.plot(X1, Y1)
plt.title("Part 1.")
plt.show()
plt.plot(X2, Y2)
plt.title("Part 2.")
plt.show()
```
